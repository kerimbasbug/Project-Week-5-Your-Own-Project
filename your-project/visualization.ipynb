```python
import matplotlib.pyplot as plt
 
# Make data: I have 3 groups and 7 subgroups
group_names=['Delay', 'Cancel&Divert']
group_size=[17, 1.73]
subgroup_names=['Carrier', "NAS", "Weather", "Aircraft Late",'Cancelled', 'Diverted']
subgroup_size=[5.19, 5.72, 0.58, 6.6, 1.61, 0.24]
 
# Create colors
a, b, c=[plt.cm.Blues, plt.cm.Reds, plt.cm.Greens]
 
# First Ring (outside)
fig, ax = plt.subplots()
ax.axis('equal')
mypie, _ = ax.pie(group_size, radius=2.7, labels=group_names, colors=[a(0.7), c(0.6), c(0.6)] )
plt.setp( mypie, width=0.6, edgecolor='white')
 
# Second Ring (Inside)
mypie2, _ = ax.pie(subgroup_size, radius=2.4-0.3, labels=subgroup_names, labeldistance=0.75, colors=[a(0.6), a(0.5), a(0.4), a(0.3), c(0.5), c(0.4)])
plt.setp( mypie2, width=0.8, edgecolor='white')
plt.margins(0,0)
 
# show it
plt.show()
```
