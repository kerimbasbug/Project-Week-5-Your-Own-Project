```python
import pandas as pd
import numpy as np
```

```python
jm = pd.read_csv("lax_jfk.csv",  skiprows=5)
```

```python
jm.head()
```

```python
index = jm[jm['Carriers']=="Carriers"].index.values.astype(int)[0]
```

```python
df1 = jm.head(index)
```

```python
df1 = df1[["Carriers", "Average Airborne Time (minutes)", "Average Taxi-Out Time (minutes)", "Average Taxi-In Time (minutes)"]]
```

```python
df1 = df1[1:]
```

```python
df1 = df1.reset_index(drop=True).reset_index()
df1
```

```python
df2 = pd.read_csv("jan16_jan19All.csv")
```

```python
df2 = df2.loc[df2["airport"]== "LAX"].groupby(["carrier"]).sum().reset_index()
```

```python
car_list = list(df1.Carriers.unique())
car_list
```

```python
df2 = df2.loc[df2["carrier"].apply(lambda x: x in car_list)]
```

```python
df2 = df2.drop(["year", " month"," arr_delay", "arr_del15", "carrier_ct", " weather_ct", 'nas_ct', 'security_ct', 'late_aircraft_ct',
       'arr_cancelled', 'arr_diverted', "Unnamed: 21"], axis=1)
df2
```

```python
df2["avg_car"] = df2[" carrier_delay"]/df2["arr_flights"]
df2["avg_wea"] = df2["weather_delay"]/df2["arr_flights"]
df2["avg_nas"] = df2["nas_delay"]/df2["arr_flights"]
df2["avg_sec"] = df2["security_delay"]/df2["arr_flights"]
df2["avg_lad"] = df2["late_aircraft_delay"]/df2["arr_flights"]
df2.head()
```

```python
df2 = df2.drop(["arr_flights", " carrier_delay", "weather_delay", "nas_delay", "security_delay", "late_aircraft_delay"], axis=1)
```

```python
df2 = df2.reset_index(drop=True).reset_index()
```

```python
df2
```

```python
df1
```

```python
res = pd.merge(df1, df2, on="index")
```

```python
res
```

```python
res = res.drop(["carrier"], axis=1)
res.columns
```

```python
res = res.rename(columns={"Average Airborne Time (minutes)": "avg_dist", "Average Taxi-Out Time (minutes)": "avg_taxi_out",
                    "Average Taxi-In Time (minutes)": "avg_taxi_in"})
```

```python
res["avg_dist"] = res["avg_dist"].astype(float)
res["avg_taxi_out"] = res["avg_taxi_out"].astype(float)
res["avg_taxi_in"] = res["avg_taxi_in"].astype(float)
```

```python
res.info()
res
```

```python
for i in range(res["index"].max()+1):
    dk = 10.57 - (7.95)*(10**-4)*res["avg_dist"][i] - (0.4392)*res["avg_taxi_in"][i] - (0.557)*res["avg_taxi_out"][i] + (0.979)*res["avg_car"][i] + (0.991)*res["avg_wea"][i] + (0.908)*res["avg_nas"][i] + (1.119)*res["avg_sec"][i] + (0.881)*res["avg_lad"][i]
    print(res["Carriers"][i], dk*5)
```
