```python
df = df.drop(["year", " month"], axis=1)
```

```python
df = df.drop(lst, axis=1)
```

```python
df
```

Calculating the ratios of delay types to all flights

```python
df["rarr_del15"] = 100 *df["arr_del15"] / df["arr_flights"]
df["rcarrier_ct"] = 100 *df["carrier_ct"] / df["arr_flights"]
df["rweather_ct"] = 100 *df[" weather_ct"] / df["arr_flights"]
df["rnas_ct"] = 100 *df["nas_ct"] / df["arr_flights"]
df["rsecurity_ct"] = 100 *df["security_ct"] / df["arr_flights"]
```

```python
df["rarr_cancelled"] = 100 *df["arr_cancelled"] / df["arr_flights"]
df["rarr_diverted"] = 100 *df["arr_diverted"] / df["arr_flights"]
```

```python
df.head()
```

```python
lst = ['arr_del15', 'carrier_ct',
       ' weather_ct', 'nas_ct', 'security_ct', 'late_aircraft_ct',
       'arr_cancelled', 'arr_diverted']
df = df.drop(lst, axis=1)
```

```python
df.head()
```

```python
df["sum"] = 100-df["rarr_del15"]-df["rcarrier_ct"]-df["rweather_ct"]-df["rnas_ct"]-df["rsecurity_ct"]-df["rarr_cancelled"]-df["rarr_diverted"]
```

```python
df.sort_values(by="sum", ascending=False).reset_index()
```

```python
df.to_csv("performance.csv")
```

```python
# sample points 
X = df["avg_age"]
Y = df["sum"]

# solve for a and b
def best_fit(X, Y):

    xbar = sum(X)/len(X)
    ybar = sum(Y)/len(Y)
    n = len(X) # or len(Y)

    numer = sum([xi*yi for xi,yi in zip(X, Y)]) - n * xbar * ybar
    denum = sum([xi**2 for xi in X]) - n * xbar**2

    b = numer / denum
    a = ybar - b * xbar

   # print('best fit line:\ny = {:.2f} + {:.2f}x'.format(a, b))
    return a, b

# solution
a, b = best_fit(X, Y)
#best fit line:
#y = 0.80 + 0.92x

# plot points and fit line
import matplotlib.pyplot as plt
plt.scatter(X, Y)
yfit = [a + b * xi for xi in X]
plt.plot(X, yfit)
```
